```python
mean_population = df.miles.mean()
std_population = df.miles.std()
mean_population,std_population
```

```python
sample_size = 50
sample_df = df.sample(sample_size , random_state=100)
sample_df
```

```python
mean_sample = sample_df.miles.mean()
std_sample = sample_df.miles.std()
mean_sample , std_sample
```

```python
z_score_95_confidence = st.norm.ppf(0.975)
z_score_95_confidence
```

```python
margin_of_error = z_score_95_confidence*(std_population/np.sqrt(sample_size))
margin_of_error
```

```python
lower = mean_sample - margin_of_error
upper = mean_sample + margin_of_error
lower , upper
```
